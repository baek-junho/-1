```python
import seaborn as sns
```

```python
raw = sns.load_dataset('tips')
```

```python
raw.head()
```

```python
raw.info()
#  실수형  카테고리형 정수형 
```

```python
# sns.relplot() 는 두 데이터간(수치,수치)의 관계를 시각화 할수 있다 
# kind 기본은 스케터
sns.relplot(data=raw, x = 'tip', y = 'total_bill' )
```

```python
sns.relplot(data=raw, x='tip', y='total_bill', kind='line')
```

```python
sns.relplot(data=raw, x='tip', y='total_bill', hue='sex')
```

```python
sns.jointplot(data=raw, x='tip', y='total_bill')
# 축에대한 데이터분포가 히스토그램으로 나타남
```

```python
sns.jointplot(data=raw, x='tip', y='total_bill', kind='kde')
# 밀도함수로 표현 많을수록 진하게 
```

```python
sns.jointplot(data=raw, x= 'tip', y='total_bill', kind='regg')
# 두 관계에 대한 회귀식 확인가능
```

```python
sns.jointplot(data=raw, x='tip', y='total_bill', kind='hex')
```

```python
#pairplot모든 수치형 컬럼에서 두 컬럼씩 관계를 시각화 가능
sns.pairplot(data=raw)
```

```python
sns.pairplot(data=raw, hue='sex')
```

수치/카테고리형

```python
sns.boxplot(data=raw, x='day', y='tip')
# 특정 기준에 따른 데이터분포를 살펴보기에 용의한 그래프 분포는 표시하지만 해당하는 갯수는 표기x 
# 데이터 개수가 다른 값을 비교하기에는 문제가 있을수 있다 그래서 스왐플롯 사용
```

```python
sns.boxplot(data=raw, x='day', y='tip', hue='smoker')
```

```python
sns.swarmplot(data=raw, x='day', y='tip')
```

```python
sns.swarmplot(data=raw, x='day', y='tip', hue='smoker', dodge=True)
# dodge 흡연여부에 따라서 그래프 분리
```

```python
sns.boxplot(data=raw, x='day', y='tip', hue='smoker')
sns.swarmplot(data=raw, x='day', y='tip', hue='smoker', dodge=True)
```

```python
sns.boxplot(data=raw, x='size', y='tip', hue='sex')
```

```python
# barplot의 경우 특정 값에 대한 개수를 살펴볼수 있다
sns.barplot(data=raw, x='size', y='tip')
```

수치형/카테고리형/카테고리형

```python
df =raw.pivot_table(index='day', columns='size', values='tip', aggfunc='mean')
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot=True) # annot true로 하면 수치나타남 fmt(포멧팅) = 소수점자리표시
```

```python
sns.heatmap(df, annot=True, cmap='Pastel1') 
```
